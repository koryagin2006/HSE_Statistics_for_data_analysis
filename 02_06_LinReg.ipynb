# Линейная регрессия

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as st
import statsmodels.api as sm
import numpy as np

bikes = pd.read_pickle('Data/BikesDataImputed.pkl')
bikes.head(n=3)
```

## Проведем аггрегацию данных по неделям

```python
bikes_weeks = bikes \
    .groupby(by=bikes['Date'].dt.isocalendar().week) \
    .agg({'Rental Count': 'sum', 
          'Temperature': 'mean', 
          'Humidity': 'mean', 
          'Wind speed': 'mean'})
```

## Проверка условий для построения модели

### Необходимо проверить отсутстивие мультиколлинеарности

```python
plt.figure(figsize=(6, 6))
p = sns.heatmap(data=bikes_weeks.corr(), 
                annot=True, 
                cmap='RdYlGn', 
                vmin=-1, vmax=1)
```

### Наличие линейной взаимосвязи

```python
sns.pairplot(bikes_weeks);
```

## Построение модели

### scipy.stats.

NB! Возможно только для одного предиктора

```python
X = bikes_weeks['Temperature']
y = bikes_weeks['Rental Count']

res = st.linregress(x=X, y=y)
print(res)

# slope - угол наклона
# intercept - значение для X=0
# rvalue - это сама r, перед r_квадрат
# pvalue - стат значимость зависимости, 
# stderr - стандартная ошибка
```

```python
# Выведем метрику r-квадрат

r_sq = res.rvalue ** 2
print('r_sq =', r_sq)

# Моделт на 71% объясняет диперсию (разброс) целевой переменной
```

```python
# Простроим график
y_pred = res.slope * X + res.intercept

plt.figure(figsize=(10, 5))
plt.scatter(X, y, label='Оригинальные данные')
plt.plot(X, y_pred, color='r', label='Линия регрессии')
plt.legend()
plt.show()
```

### statsmodels.api

#### Сначала построим модель на 1 предикторе

```python
X_simple = bikes_weeks['Temperature']  # создаем отдельную переменную
X_simple_constnt = sm.add_constant(X_simple)
y = bikes_weeks['Rental Count']

model = sm.OLS(y, X_simple_constnt)  # OLS (Ordinary Least Squares) - метод наименьших квадратов
res_simple = model.fit()
res_simple.summary()
```

- `Method: Least Squares`	 - описание метода
- `R-squared:	0.715` - $R^2$
- `const 6.174e+04` - интерсепт
- `Temperature	4309.3206` - коэссфициент (угол наклона)

#### Модель на нескольких предикторах

```python
X_mult = sm.add_constant(bikes_weeks.drop('Rental Count', axis=1))

model = sm.OLS(y, X_mult)
res_mult = model.fit()
res_mult.summary()
```

 - const	1.756e+05 - интерсепт
 - Temperature	4606.1894 - коэффициент перед переменой
 - Humidity	-1228.9486
 - Wind speed	-2.675e+04

```python
import statsmodels.stats.api as sms
sns.set_style('darkgrid')
sns.mpl.rcParams['figure.figsize'] = (15.0, 9.0)

def linearity_test(model, y):

    fitted_vals = model.predict()
    resids = model.resid

    fig, ax = plt.subplots(1,2)
    
    sns.regplot(x=fitted_vals, y=y, lowess=True, ax=ax[0], line_kws={'color': 'red'})
    ax[0].set_title('Observed vs. Predicted Values', fontsize=16)
    ax[0].set(xlabel='Predicted', ylabel='Observed')

    sns.regplot(x=fitted_vals, y=resids, lowess=True, ax=ax[1], line_kws={'color': 'red'})
    ax[1].set_title('Residuals vs. Predicted Values', fontsize=16)
    ax[1].set(xlabel='Predicted', ylabel='Residuals')
    
```

```python
linearity_test(res_simple, y)    
```

```python
linearity_test(res_mult, y)    
```
