# Визуализация распределений

### Импорт библиотек

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Графики

```python
N = 10 ** 4
np.random.seed(42)


def plot_comparison(x, title):
    """
    Построение трех графиков сразу
    x: Data
    title: Plot title
    """
    fig, ax = plt.subplots(nrows=3, ncols=1, sharex=True, figsize=(15, 10))

    sns.histplot(x=x, ax=ax[0], kde=True), ax[0].set_title('Histogram + KDE')  # KDE - линия оценки ядерной плотности
    sns.boxplot(x=x, ax=ax[1]), ax[1].set_title('Boxplot')
    sns.violinplot(x=x, ax=ax[2]), ax[2].set_title('Violin plot')

    fig.suptitle(title, fontsize=16)
    plt.show()
```

### Страндартное распределение

```python
sample_gaussian = np.random.normal(size=N)
plot_comparison(x=sample_gaussian, title='Standard Normal Distribution')
```

### Логарифмическое распределение

```python
sample_lognormal = np.random.lognormal(size=N)
plot_comparison(x=sample_lognormal, title='Log-normal Distribution')
```

### Бимодальное распределение

```python
sample_bimodal = np.concatenate([np.random.normal(loc=-2, scale=2, size=int(N / 2)),
                                 np.random.normal(loc=3, scale=1, size=int(N / 2))])
plot_comparison(x=sample_bimodal, title='Mixture of Gaussians - bimodal')
```
